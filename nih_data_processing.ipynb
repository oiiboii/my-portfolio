## Acoustic Myography (AMG) Feature Analysis

Analyzing AMG signals requires a set of features that can represent the muscle's complex activity. Here's a list of essential features and their significance:

1. **Mean Amplitude**:
    - **Description**: Represents the average amplitude of the signal over a specific time frame.
    - **Mathematical Interpretation**: \( \text{Mean Amplitude} = \frac{1}{N}\sum_{i=1}^{N} x_i \) where \( x_i \) is the signal amplitude at the ith sample, and N is the total number of samples.
  
2. **Standard Deviation**:
    - **Description**: Measures the amount of variation or dispersion in the AMG signal.
    - **Mathematical Interpretation**: \( \sigma = \sqrt{\frac{1}{N}\sum_{i=1}^{N} (x_i - \mu)^2} \) where \( \mu \) is the mean amplitude.
  
3. **Frequency Domain Features**:
    - **Description**: These features are derived from the Fourier transform of the signal, highlighting the signal's frequency content.
        - **Peak Frequency**: The frequency with the highest magnitude.
        - **Median Frequency**: The frequency below which 50% of the total power is contained.
  
4. **Time Domain Features**:
    - **Description**: Represents characteristics of the signal over time.
        - **Zero Crossings**: Times the signal crosses the zero line.
        - **Waveform Length**: Total length of the waveform over a time segment.
  
5. **Modulation Signal**:
    - **Description**: Represents the difference between the upper and lower envelopes, providing insights into the muscle's dynamic activity.
  
6. **Concentric and Eccentric Duration**:
    - **Description**: Measures the time duration of concentric (muscle shortening) and eccentric (muscle lengthening) contractions.
  
7. **Force Metrics**:
    - **Description**: Metrics derived from the force exerted during contractions, representing muscle strength.
        - **Total Force**: Cumulative force over time.
        - **Average Force per Second**: Mean force exerted per second.
  
8. **Total Load**:
    - **Description**: Represents the cumulative load experienced during different contraction phases, essential for understanding muscle workload.
    - **Mathematical Interpretation**: Product of force and duration.

9. **Firing Rate**:
    - **Description**: Indicates the rate at which muscle fibers are activated.
  
10. **Fatigue Score**:
    - **Description**: A metric that provides insights into the muscle's fatigue level over time, crucial for training and rehabilitation purposes.

# Acoustic Myography: Understanding Muscle Activity through Sound Analysis

In the realm of sports medicine, decoding muscle activity is a key pursuit. Acoustic myography, a promising field, focuses on interpreting muscle-generated sound to glean insights. This proposal harnesses acoustic myography's potential by utilizing modern technology to capture and analyze muscle-generated sounds.

## Introduction

Acoustic myography offers a non-invasive window into muscle dynamics. The core concept is that muscles emit sound patterns that can be harnessed for insights into muscle performance, fatigue, and overall health.

## Data Acquisition and Preparation

We use BioBeat's prototype to record calf contractions. Data includes sound amplitude and frequency sample rate. Libraries imported for this task are: `pandas`, `numpy`, `scipy`, `matplotlib`, and `pywt`.

## Signal Processing and Feature Engineering

Signal processing sharpens crucial features of raw muscle-generated sounds for precise analysis.

1. **Data Normalization:** Normalize data to maintain consistent scales.

2. **Fourier Transform:** The Fourier transform helps unveil frequency components of the sound signal.

3. **Signal Envelope & Peak Detection:** extract envelope and detect peaks to understand intensity and rhythm of muscle contractions.

4. **Signal Enhancement and Modulation:** Enhancements clarify the signal, while modulation helps assess amplitude variations creates an amplitude-modulated signal based on the difference between the upper and lower envelopes of our myography data. This modulated signal gives a high-frequency representation of the muscle activity fluctuations.

## Visualization

Visualizations illuminate the intricacies of processed signals, offering intuitive insights into muscle activity dynamics.

```python
# Importing necessary libraries
import pandas as pd
import numpy as np
from scipy.fft import fft, ifft
from scipy.signal import hilbert, savgol_filter, butter, filtfilt, find_peaks, medfilt
from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt
import pywt

# Utility functions
def butter_bandpass_filter(data, lowcut, highcut, fs, order=3):
    nyquist = 0.5 * fs
    low = lowcut / nyquist
    high = highcut / nyquist
    b, a = butter(order, [low, high], btype='band')
    y = filtfilt(b, a, data)
    return y

def moving_average(data, window_size):
    return np.convolve(data, np.ones(window_size)/window_size, mode='valid')

# Data loading and preliminary processing
csv_path = "/content/drive/MyDrive/STTR/data.csv"
df = pd.read_csv(csv_path)
df.rename(columns={'FS ': 'FS', 'Data 2 ': 'Data 2'}, inplace=True)
df['FS'].fillna(44100.0, inplace=True)
amplitude = df['Data 1'].values
fs = df['FS'][0]

# Normalizing the amplitude
amplitude = (amplitude - np.min(amplitude)) / (np.max(amplitude) - np.min(amplitude))

# Signal processing
normalized_data = (df['Data 1'] - df['Data 1'].min()) / (df['Data 1'].max() - df['Data 1'].min())
centered_normalized_data = normalized_data - np.mean(normalized_data)
f_transform = fft(amplitude)
frequencies = np.fft.fftfreq(len(f_transform), 1/fs)
positive_freq_idxs = np.where(frequencies > 0)
f_transform = f_transform[positive_freq_idxs]
frequencies = frequencies[positive_freq_idxs]
filtered_transform = np.where((frequencies >= 0.1) & (frequencies <= 200), f_transform, 0)
filtered_amplitude = np.real(ifft(filtered_transform))

# Applying Butterworth bandpass filter
filtered_data = butter_bandpass_filter(df['Data 1'], 5, 200, fs, order=3)

# Extracting peaks and valleys for envelope estimation
N=10
peaks, _ = find_peaks(filtered_data, distance=fs//N)
valleys, _ = find_peaks(-filtered_data, distance=fs//N)
if len(peaks) >= 2 and len(valleys) >= 2:
    upper_interp = CubicSpline(df['Time 1'][peaks], filtered_data[peaks], bc_type='natural')
    lower_interp = CubicSpline(df['Time 1'][valleys], filtered_data[valleys], bc_type='natural')
    upper_envelope = upper_interp(df['Time 1'])
    lower_envelope = lower_interp(df['Time 1'])

# Signal enhancement techniques
filtered_data_enhanced = medfilt(filtered_data, kernel_size=5)
filtered_data_enhanced = savgol_filter(filtered_data_enhanced, window_length=101, polyorder=3)

# Visualization
fig, axes = plt.subplots(nrows=4, figsize=(12, 14))
axes[0].plot(df['Time 1'], centered_normalized_data, label='Original', alpha=0.5)
axes[0].plot(df['Time 1'], filtered_data, label='Filtered', alpha=0.5)
axes[0].set_title("Original VS Filtered")
axes[0].legend()

Y = fft(filtered_data)
axes[1].plot(frequencies, np.abs(f_transform), label='Original FFT')
axes[1].plot(frequencies, np.abs(Y[:len(frequencies)]), label='Filtered FFT', alpha=0.5)
axes[1].set_title("Original VS Filtered FFT")
axes[1].set_xlim(0, 500)
axes[1].legend()

axes[2].plot(df['Time 1'], upper_envelope, label='Upper Envelope')
axes[2].plot(df['Time 1'], lower_envelope, label='Lower Envelope')
axes[2].plot(df['Time 1'][peaks], filtered_data[peaks], 'ro', label='Peaks')
axes[2].plot(df['Time 1'][valleys], filtered_data[valleys], 'go', label='Valleys')
axes[2].set_title("Envelope with Peak Detection")
axes[2].legend()

# 1. Compute the modulation signal
modulation_signal = upper_envelope - lower_envelope

# 2. Generate a high-frequency sine wave (e.g., 100 Hz)
carrier_frequency = 1000  # Adjust as needed
sine_wave = np.sin(2 * np.pi * carrier_frequency * df['Time 1'])

# 3. Multiply sine wave by the modulation signal to get the modulated sine wave
modulated_sine = modulation_signal * sine_wave

# Plot the modulated sine wave
plt.plot(df['Time 1'], modulated_sine, label='Modulated Sine')
plt.legend()
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.title('Amplitude Modulated Sine Wave')

plt.tight_layout()
plt.show()
```

## Concentric Vs Eccentric Motion
The given analysis aims to distinguish between concentric (muscle shortening) and eccentric (muscle lengthening) muscle activities. To achieve this:

1. **Differential Envelope Creation**: We subtract the lower envelope from the upper envelope, generating a differential envelope. The differential envelope's magnitude indicates the muscle's contraction strength.

2. **Smoothing the Differential Envelope**: This step is essential to remove noise and provide a clear signal to differentiate the muscle activities.

3. **Setting Thresholds and Minimum Duration**: A threshold based on the mean and standard deviation of the smoothed differential envelope is established. This threshold is vital in distinguishing between concentric and eccentric motions. The minimum duration helps in excluding transient changes that don't last long enough to be of significance.

4. **Visualization and Region Identification**: The plotted graph depicts the enhanced muscle signal and the smoothed differential envelope. Regions of concentric and eccentric activity are color-highlighted for clarity. Red indicates concentric actions, while green shows eccentric activities. This visualization provides an intuitive way to understand muscle activity throughout the recorded period.

```python
from scipy.ndimage import gaussian_filter1d
from matplotlib.lines import Line2D

# Create the differential envelope signal from the upper and lower envelopes
diff_envelope = upper_envelope - lower_envelope

# Apply Gaussian filter to smooth the differential envelope for noise reduction
smooth_diff_envelope = gaussian_filter1d(diff_envelope, sigma=5)

# Define threshold based on the mean and standard deviation of the smoothed differential envelope
# This threshold is used to distinguish between concentric and eccentric motions
threshold = np.mean(smooth_diff_envelope) + 0.7 * np.std(smooth_diff_envelope)

# Set a minimum duration to avoid identifying very short motions (adjustable)
min_duration = 0.45 * fs  # half-second duration

regions = []
start = 0
current_motion = 'concentric' if smooth_diff_envelope[0] < threshold else 'eccentric'

# Identify regions of concentric and eccentric motions
for i in range(1, len(smooth_diff_envelope)):
    # Change in motion detected
    if (smooth_diff_envelope[i] < threshold and current_motion == 'eccentric') or \
       (smooth_diff_envelope[i] > threshold and current_motion == 'concentric'):
        if i - start < min_duration:  # If the detected region is too short, skip
            continue
        regions.append((current_motion, (start, i)))
        start = i
        current_motion = 'concentric' if smooth_diff_envelope[i] < threshold else 'eccentric'

# Visualization
plt.figure(figsize=(15, 7))

# Plot enhanced signal and smoothed differential envelope
line1, = plt.plot(df['Time 1'], filtered_data_enhanced, label="Enhanced Signal", color="black")
line2, = plt.plot(df['Time 1'], smooth_diff_envelope, label="Smoothed Differential Envelope", color="blue", alpha=0.7)

# Custom legend for regions of motion and signals
custom_legend = [
    Line2D([0], [0], color='red', alpha=0.5, lw=10, label='Concentric region'),
    Line2D([0], [0], color='green', alpha=0.5, lw=10, label='Eccentric region'),
    Line2D([0], [0], color=line1.get_color(), label=line1.get_label()),
    Line2D([0], [0], color=line2.get_color(), label=line2.get_label())
]

# Highlight the identified regions of concentric and eccentric motions
for motion, window in regions:
    color = 'red' if motion == "concentric" else 'green'
    plt.axvspan(df['Time 1'][window[0]], df['Time 1'][window[1]], color=color, alpha=0.5)

# Provide title and axis labels
plt.title("Enhanced Signal with Identified Regions of Motion")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.legend(handles=custom_legend, loc='upper right')
plt.show()

# Display some statistics
print(f"Total number of identified regions: {len(regions)}")
print(f"Number of concentric regions: {len([r for r in regions if r[0] == 'concentric'])}")
print(f"Number of eccentric regions: {len([r for r in regions if r[0] == 'eccentric'])}")
```

## Features Extraction and Regions Analysis**

In this section, we expand our analysis of the acoustic myography signal by extracting key features from the identified regions of concentric and eccentric muscle activity. Extracting these features provides deeper insights into muscle behavior and dynamics, making it possible to better understand muscle performance and potential areas of improvement or concern.

### **Analysis Explanation:**

**1. Constructing Regions DataFrame:**
After identifying regions of concentric and eccentric motions, the segments corresponding to these regions are stored in a DataFrame. This includes the type of motion, the time segments, and the amplitude of the signal during these segments.

**2. Features Extraction:**
From each identified region, we extract:
- **Mean Amplitude**: The average amplitude of the signal during that region, representing the intensity of the muscle contraction.
- **Duration**: The total duration of the region, indicating how long the muscle was in that specific motion.
- **Contractions Count**: Simply a count to aid in the aggregation of total contractions.

**3. Concentric and Eccentric Segmentation:**
We classify the regions based on motion type to provide individualized insights for concentric and eccentric muscle activity.

**4. Statistical Analysis:**
For both concentric and eccentric motions, we aggregate:
- **Total Time**: The sum of the duration of all regions, revealing how long the muscle performed each type of motion.
- **Average Amplitude**: The mean amplitude across all regions, showcasing the average intensity of muscle contractions.
- **Total Contractions**: The total number of identified regions, indicating the number of times the muscle switched to that specific motion.

```python
import pandas as pd
import numpy as np

# Create a DataFrame to store the details of each identified contraction region
df_regions = pd.DataFrame(columns=["Motion", "Time_Segment", "Signal_Segment"])

# Populate the DataFrame with motion type, corresponding time segments, and signal segments
for motion, (start, end) in regions:
    new_row = pd.DataFrame({
        "Motion": [motion],
        "Time_Segment": [df['Time 1'][start:end].values],
        "Signal_Segment": [smooth_diff_envelope[start:end]]
    })
    df_regions = pd.concat([df_regions, new_row], ignore_index=True)

# Calculate additional features for each contraction region
df_regions['Mean_Amplitude'] = df_regions['Signal_Segment'].apply(np.mean)
df_regions['Duration'] = df_regions['Time_Segment'].apply(lambda x: x[-1] - x[0])
df_regions['Contractions_Count'] = 1

# Split the data into concentric and eccentric for targeted analysis
df_concentric = df_regions[df_regions["Motion"] == "concentric"]
df_eccentric = df_regions[df_regions["Motion"] == "eccentric"]

# Statistical Analysis
total_time_concentric = df_concentric['Duration'].sum()
total_time_eccentric = df_eccentric['Duration'].sum()
avg_amplitude_concentric = df_concentric['Mean_Amplitude'].mean()
avg_amplitude_eccentric = df_eccentric['Mean_Amplitude'].mean()
total_concentric_contractions = df_concentric['Contractions_Count'].sum()
total_eccentric_contractions = df_eccentric['Contractions_Count'].sum()

# Compile the analysis results into a DataFrame for easy viewing
analysis_df = pd.DataFrame({
    'Metrics': ['Total Time (Concentric)', 'Total Time (Eccentric)', 'Average Amplitude (Concentric)',
                'Average Amplitude (Eccentric)', 'Total Concentric Contractions', 'Total Eccentric Contractions'],
    'Values': [total_time_concentric, total_time_eccentric, avg_amplitude_concentric, avg_amplitude_eccentric,
               total_concentric_contractions, total_eccentric_contractions]
})

analysis_df
```

### Wavelet Transforms in Acoustic Myography (AMG)

Acoustic Myography (AMG) focuses on analyzing the sounds produced by muscle contractions. Wavelet Transforms, particularly the Continuous Wavelet Transform (CWT), can provide a time-frequency representation of these sounds, making them instrumental for pattern recognition.

#### **Benefits for AMG**:

1. **Pattern Recognition**: Using the image-like representation of CWT, patterns corresponding to specific muscle activities or anomalies can be discerned.
   
2. **Anomaly Detection**: The CWT can help spot anomalies or abnormalities in muscle sounds, which might indicate muscle dysfunction or other issues.
   
3. **Data Compression**: As AMG can produce vast amounts of data, compressing this information is essential. By turning the signal into an "image" using CWT and subsequently compressing it (e.g., using kernel-based operations), we retain significant patterns while reducing data size.

#### **Practical Implementation**:

To integrate with future research on AMG, a concise pipeline could be:

- Apply CWT on AMG signals to convert them into image-like structures.
- Compress these images to retain significant patterns and reduce data complexity.
- Utilize computer vision techniques (traditional or machine learning-based) to identify and categorize muscle activities or anomalies.
